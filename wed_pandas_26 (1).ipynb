# Slicing with conditions (boolean indexing)
## Basic condition.
#### Vehicles with mpg greater than 18.0
``` df[df['mpg'] > 18.0] ```
This will return all columns in the row where mpg is greater than 18.0

#### How it works.
step 1: df['mpg'] > 18.0 , returns a boolean value.

step 2: Boolean series to filter, only outputs on true rows.

````python
(df[~(df['mpg'] > 18.0 )]).describe()
````

````python
df.describe()
````

### Multiple conditions.
#### AND (&)
``` df[(df['mpg'] > 18.0) & (df['cylinders'] == 6)]```

#### OR (|)

#### NOT (~)


````python
(df[(df['mpg'] > 18.0) & (df['cylinders'] == 6)]).describe()
````

````python
df.head()
````

````python
# Using string conditions eg contains
(df[df['car name'].str.contains('chevrolet')]).describe(include='all')
````

### adding columns based on a condition.

)

### Handling missing data.
Finding or detecting missing values.

``` df.isnull() ```
returns a dataframe of True/False, True where we have a missing value.

``` df.isnull().sum() ```
returns count of missing values per column. Takes True as 1 and False as 0.

``` df.isnull().sum().sum() ``` 
returns the total missing values in a given dataframe.

``` (df.isnull().sum().sum())/len(df)```
returns percentage missing.


````python
# checking for missing values.
(df.isnull().sum().sum())/len(df) #
````

## Handling duplicates
### Detecting duplicates.
``` df.duplicated() ```
returns a boolean series, True/ False.

``` df[df.duplicated()] ```
shows the duplicated rows

````python
df_duplicate = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Alice', 'Charlie'],
    'Age': [25, 30, 25, 35],
    'city': ['Nairobi', 'Kisumu', 'Nairobi', 'Mombasa']
})
````

````python
print("Before:", len(df_duplicate))
df_duplicate = df_duplicate.drop_duplicates()
print("After: ", len(df_duplicate))
````

````python
df_duplicate
````

````python
df_duplicate.drop_duplicates(keep=False)
````
